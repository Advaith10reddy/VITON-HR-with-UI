```python
import os
import shutil
import gradio as gr
from PIL import Image

def virtual_try_on(original_image, cloth_image):
    # Save uploaded images to static folder
    input_dir = 'static'
    original_path = os.path.join(input_dir, 'origin_web.jpg')
    cloth_path = os.path.join(input_dir, 'cloth_web.jpg')

    original_image.save(original_path)
    cloth_image.save(cloth_path)

    # Run your existing code
    !python main.py

    # Display the final result
    final_image_path = "./static/finalimg.png"
    final_image = Image.open(final_image_path)
    return final_image
```

```python
# Define the Gradio interface
iface = gr.Interface(fn=virtual_try_on,
                     inputs=[gr.Image(type="pil", label="Upload Your Original Image"),
                             gr.Image(type="pil", label="Upload Your Cloth Image")],
                     outputs=gr.Image(type="pil", label="Final Result"))

# Launch the Gradio interface
iface.launch(share = False)
```
